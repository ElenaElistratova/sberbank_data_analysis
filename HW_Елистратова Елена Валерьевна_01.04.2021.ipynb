```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('./data/titanic/train.csv')
```

```python
data.head()
```

```python
data = data.set_index('PassengerId')
```

```python
data = data[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked']]
```

```python
data.head()
```

```python
data = data.fillna(0)
data.head()
```

```python
#1 Больше детей выжило в 1 классе, чем в других
```

```python
data1 = data[data['Age']<18]
data1.head()
```

```python
a = data1.groupby(['Pclass'])['Survived'].value_counts().to_frame()
a
```

```python
#Вывод: в 1 и 2 классах выжило одинаковое количество детей; в 3-м классе выжило только 29% от общего количества.
```

```python
#2 Больше всего выжило женщин, у которых 2 и менее родственников.
```

```python
data['Family']=pd.cut(data['SibSp']+data['Parch'], [0,2,5,7],include_lowest=True)
data.head()
```

```python
b = data.groupby(['Sex','Family'])['Survived'].value_counts().to_frame()
b
```

```python
#Вывод: действительно, больше всего выжило женщин, у которых 2 и менее родственников.
```

```python
#3 Анализ данных по стоимости проезда
```

```python
import numpy as np
```

```python
data.groupby(['Embarked','Pclass','Sex'])['Fare'].aggregate([np.median, np.mean])
```

```python
#Выводы: 1) В порту Q стоимость проезда для мужчин и женщин одинакова;
#2) В порту C стоимость проезда для мужчин и женщин отличается незначительно, при этом, женщины платили больше;
#3) В порту S женщины платили за проезд почти в два раза больше, что характерно для билетов во все классы
```
